```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/UzbekistonGender.csv')

plt.figure(figsize=(15, 10))
plot = sns.lineplot(markers=True, dashes=False, data=df, x="Years", y="Count", hue="Gender", style="Gender", markersize=6)
plt.title('Oliy ta\'lim muassasalarining talabalar to‘g‘risida ma\'lumot (erkaklar va ayollar sonining nisbati)')

fig = plot.get_figure() 
fig.savefig('data/OliyTalimGender.png')
```
