```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/DataSuv_Done.csv')
heatmap_data = df.pivot_table(index='City', columns='Years', values='Value')
heatmap_data = heatmap_data.sort_values(by="City", key=lambda x: x.str.len())

# Heat map
plt.figure(figsize=(20, 10))
plot = sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g', linewidths=0.5) 
plt.title('Hududlar bo`yicha kvartira (uy)larning ichimlik suvi bilan ta`minlanganligi yillar bo\'yicha! (Yil oxiriga, jami kvartira (uy)larga nisbatan, % da )')
# plt.show() 

fig = plot.get_figure() 
fig.savefig('data/suv.png')
 
```
